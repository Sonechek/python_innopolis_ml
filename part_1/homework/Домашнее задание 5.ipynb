Для тестирования API нового сервиса валидации адресов им необходимо получить списки реальных адресов и геокоординат которыми пользуются клиенты компании.\

Файл durty_data.csv прилагается к письму.

От вас требуется выделить из файла только похожие на адреса и гео-координаты строки. Убрать дубликаты и кавычки, а также разложить результаты на 2 разных листа Excel файла: строковые адреса на один, а гео-координаты на другой.

Результатом работы по вашему алгоритму должен стать xlsx файл, в котором на разных листах лежат все распознанные адреса и геокоординаты.

#### *Решение*

```python
import pandas as pd
import re

# Получение данных из файла
# Предполагается, что файл "durty_data.csv" находится в той же директории, что и этот скрипт
# Если файл в другом месте, укажите полный путь к файлу
data = pd.read_csv("durty_data.csv", header=None)

# Очистка строк от кавычек и пробелов
data[0] = data[0].str.replace('"', '').str.strip()

# Удаление дубликатов 
data = data.drop_duplicates()

# Регулярные выражения для фильтрации
geo_reg = r'[\d.-]+,\s*[\d.-]+'
address_reg = r'[а-яА-Я]+(?:\s+[а-яА-Я]+)+'

# Фильтрация данных
# geo_data - строки с координатами
# address_data - строки с адресами
geo_data = data[data[0].str.match(geo_reg)]
address_data = data[data[0].str.contains(address_reg, na=False)]

# Сохранение в Excel файл с двумя листами
# Первый лист - адреса, второй - координаты
with pd.ExcelWriter("filtered_addresses.xlsx") as writer:
    address_data.to_excel(writer, sheet_name="Адреса", index=False, header=False)
    geo_data.to_excel(writer, sheet_name="Координаты", index=False, header=False)
```
