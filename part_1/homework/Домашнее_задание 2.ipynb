### Задание 1: Анализ данных о студентах с использованием NumPy


**Описание:**\
У вас есть данные о студентах, включающие оценки по различным предметам. Используйте библиотеку NumPy для анализа этих данных.

```python
students_data = {
    'math': [85, 78, 92, 70, 88],
    'physics': [90, 82, 76, 85, 89],
    'informatics': [88, 92, 80, 87, 90]
}

students_data
```

**Задание:**
* Найдите среднюю оценку по каждому предмету.
* Найдите медианную оценку по каждому предмету.
* Вычислите стандартное отклонение по каждому предмету.
* Определите предмет с самой высокой средней оценкой.

```python
# Решение
import numpy as np

# Функция для расчета средней, медианной и стандартной оценки по предметам
def calculate_scores(data, metric_func):
    return {subject: metric_func(scores) for subject, scores in data.items()}

# Функция для определения предмета с самой высокой средней оценкой
def high_score_subject(average_results):
    highest_subject = max(average_results, key=average_results.get)
    return highest_subject, average_results[highest_subject]

# Расчеты
average_results = calculate_scores(students_data, np.mean)
median_results = calculate_scores(students_data, np.median)
std_results = calculate_scores(students_data, np.std)
highest_subject, highest_average = high_score_subject(average_results)

# Вывод результатов
for metric_name, results in [("Средняя оценка", average_results),
                             ("Медианная оценка", median_results),
                             ("Стандартное отклонение", std_results)]:
    print(f"{metric_name}:")
    for subject, value in results.items():
        print(f"  {subject}: {value:.2f}")
    print()

print(f"Предмет с самой высокой средней оценкой: {highest_subject} ({highest_average:.2f})")
```

### Задание 2: Анализ данных о продажах с использованием Pandas

**Описание:**\
Используйте библиотеку Pandas для анализа данных о продажах товаров. У вас есть данные о продажах товаров за последний год.

```python
sales_data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'],
    'Sales': [15000, 17000, 16000, 18000, 14000, 19000, 22000, 21000, 23000, 25000, 24000, 26000]
}
sales_data
```

**Задание:**
* Найдите общий объем продаж за год.
* Найдите средний объем продаж в месяц.
* Найдите месяц с наибольшим и наименьшим объемом продаж.
* Постройте линейный график изменения объема продаж по месяцам (Matplotlib).

```python
# Решение
import pandas as pd
import matplotlib.pyplot as plt

# Функция для выполнения расчетов
def analyze_sales(data):
    df = pd.DataFrame(data)
    
    # Общий объем продаж за год
    total_sales = df['Sales'].sum()
    
    # Средний объем продаж в месяц
    average_sales = df['Sales'].mean()
    
    # Месяц с максимальными продажами
    max_sales_month = df.loc[df['Sales'].idxmax()]
    
    # Месяц с минимальными продажами
    min_sales_month = df.loc[df['Sales'].idxmin()]
    
    return total_sales, average_sales, max_sales_month, min_sales_month, df

# Функция для построения графика
def plot_sales(df):
    df.plot(x='Month', y='Sales', kind='line', marker='o', title='Изменение объема продаж по месяцам', figsize=(8, 5))
    plt.xlabel('Месяц')
    plt.ylabel('Объем продаж')
    plt.grid(True)
    plt.show()

# Вызов функции анализа
total_sales, average_sales, max_sales_month, min_sales_month, sales_df = analyze_sales(sales_data)

# Вывод результатов
print(f"Общий объем продаж за год: {total_sales}")
print(f"Средний объем продаж в месяц: {average_sales:.2f}")
print(f"Месяц с максимальными продажами: {max_sales_month['Month']} ({max_sales_month['Sales']})")
print(f"Месяц с минимальными продажами: {min_sales_month['Month']} ({min_sales_month['Sales']})")

# Построение графика
plot_sales(sales_df)


```

### Задание 3: Задание на визуализацию данных с использованием Seaborn

**Описание:**\
Используйте готовый набор данных из библиотеки Seaborn для создания различных типов графиков. \
Ваша задача — загрузить датасет ***tips*** из готового набора Seaborn и построить несколько базовых графиков для анализа данных.

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('tips') # используйте строго данный датасет

df.head()
```

Постройте следующие графики:
* Линейный график
* Столбчатый график
* Гистограмма
* Диаграмма рассеяния

Условия:
* Параметры визуализации на ваш выбор
* Графики должны нести смысловую нагрузку

```python
# Решение
import seaborn as sns
import matplotlib.pyplot as plt

# Линейный график
plt.figure(figsize=(9, 5))
sns.lineplot(data=df, x='size', y='total_bill', marker='o')
plt.title('Сумма чека от количества')
plt.xlabel('Размер группы')
plt.ylabel('Сумма счета')
plt.grid(True)
plt.show()

# Столбчатый график
plt.figure(figsize=(9, 5))
sns.barplot(data=df, x='day', y='total_bill', hue='sex', palette=['#FFA07A', '#20B2AA'])
plt.title('Средний чек по дням недели и полу')
plt.xlabel('День')
plt.ylabel('Средняя сумма (Руб.)')
plt.legend(title='Пол')
plt.grid(True, axis='y', alpha=0.2)
plt.show()

# Гистограмма
plt.figure(figsize=(9, 5))
sns.histplot(data=df, x='total_bill', hue='smoker')
plt.title('Распределение сумм чеков по курящим и некурящим')
plt.xlabel('Сумма (Руб.)')
plt.ylabel('Количество')
plt.legend(title='Курящий', labels=['Нет', 'Да'])
plt.grid(True)
plt.show()

# Диаграмма рассеяния
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time', style='sex', markers={'Male': 'o', 'Female': 's'})
plt.title('Зависимость чаевых от суммы счета и пола')
plt.xlabel('Сумма счета')
plt.ylabel('Чаевые')
legend = plt.legend(title='Параметры', bbox_to_anchor=(1.05, 1), loc='upper left')
legend.get_texts()[1].set_text('Обед')
legend.get_texts()[2].set_text('Ужин')
legend.get_texts()[4].set_text('Муж')
legend.get_texts()[5].set_text('Жен')

plt.grid(True)
plt.tight_layout()
plt.show()

```
